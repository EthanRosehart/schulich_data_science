```python
# Probability Distribution Function .pdf function is useful when you want to know how dense the probability 
# is at a specific point on the curve, but not the cumulative probability up to that point.

sp.stats.norm.pdf(14,10,2)
```

```python
sp.stats.norm.cdf(14,10,2)
```

```python
# .ppf gives you the x-value when given a cumulative area where the given area is true
# .ppf uses the cumulative area
# .isf uses the tail area
# Can use these to find the range of the data you want to be effective
# You want to exclude the extreme values (i.e tail ends 2.5% on each side - 95% confidence)
# ppf(0.975,0,1)
# ppf(0.025,0,1)

sp.stats.norm.ppf(0.9772498680518208,10,2)
```

```python
sp.stats.norm.isf(0.9772498680518208,10,2)
```

```python
sp.stats.norm.cdf(2,0,1)
```

```python
sp.stats.norm.isf(1-0.9772498680518208,0,1)*2+10
```

```python
shopping_basket = pd.read_csv("https://raw.githubusercontent.com/EthanRosehart/schulich_data_science/main/term2/Predictive_Modeling/small_dataset.csv")
shopping_basket.head()
```

```python
from mlxtend.preprocessing import TransactionEncoder

transaction_list = list(shopping_basket["ItemsPurchased"].apply(lambda t:t.split(",")))
encoder = TransactionEncoder().fit(transaction_list)
encoded_transaction_list = encoder.transform(transaction_list)
ideal_format_table = pd.DataFrame(encoded_transaction_list, columns = encoder.columns_)
ideal_format_table.head()
```
